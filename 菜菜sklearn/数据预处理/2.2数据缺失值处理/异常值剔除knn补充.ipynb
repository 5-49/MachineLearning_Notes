```python
import pandas as pd
data = pd.read_csv(r"..\\Narrativedata.csv", index_col=0)
```

# 异常值检测

```python
data.dropna(axis=0,inplace=True)
data.head(10)
```

```python
# 编码，将标签（这里是survived）编码
from sklearn.preprocessing import LabelEncoder
data.iloc[:,-1]=LabelEncoder().fit_transform(data.iloc[:,-1]) 

from sklearn.preprocessing import OrdinalEncoder
data.iloc[:,1:-1] = OrdinalEncoder().fit_transform(data.iloc[:,1:-1])

data.head(10)
```

```python
# 接着导入EllipticEnvelope
from sklearn.covariance import EllipticEnvelope
detector = EllipticEnvelope() # 构造异常值识别器
detector.fit(data) # 拟合识别器
detector.predict(data) # 预测异常值
# -1意为异常值 
```

```python
import numpy as np

incorrect = detector.predict(data)
len(incorrect)
```

```python
for i in range(len(incorrect)):
    if(incorrect[i]==-1):
        data.iloc[[i]]=np.nan # dataframe类型的数据要这样读一行！
        # 不能像rray一样索引[i,:]

data.head(30)
```

```python
data.iloc[[0]]=np.nan
```

# k-最近邻（kNN）缺失数据的值插补

```python
# 缺失值填充过程
# 导入KNNImputer
from sklearn.impute import KNNImputer
import pandas as pd
imputer = KNNImputer(n_neighbors=2) # 这里可以设置参数n_neighbors，默认值为5
data=pd.DataFrame(imputer.fit_transform(data.values))

data
```
